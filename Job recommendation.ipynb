```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
jobs = [
    "Data Scientist with Python and Machine Learning experience",
    "Marketing Manager with SEO expertise",
    "Healthcare professional with nursing skills"
]
user_profile = "Python, Machine Learning, Data Analysis"
```

```python
# Feature Extraction
tfidf = TfidfVectorizer(stop_words='english')
job_features = tfidf.fit_transform(jobs)
user_features = tfidf.transform([user_profile])
```

```python
similarity_scores = cosine_similarity(user_features, job_features).flatten()
recommended_jobs = np.argsort(similarity_scores)[::-1]
```

```python
# Get Recommended Jobs
recommended_jobs = np.argsort(similarity_scores)[::-1]
```

```python
# Display Recommendations
for idx in recommended_jobs:
    print(f"Job: {jobs[idx]} | Score: {similarity_scores[idx]:.2f}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(jobs, similarity_scores)
plt.xlabel('Similarity Score')
plt.ylabel('Job Titles')
plt.title('Job Recommendations Based on Profile')
plt.show()
```
